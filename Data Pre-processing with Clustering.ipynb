```python
def convert_trade_status(status):
    if status == 'Buy':
        return 1
    else:
        return 0
```

```python
Bottom_Picking['Trade Status'] = Bottom_Picking['Trade Status'].apply(convert_trade_status)
Trend_Following['Trade Status'] = Trend_Following['Trade Status'].apply(convert_trade_status)
```

```python
print(Trend_Following.shape)
print(Bottom_Picking.shape)
```

```python
Bottom_Picking_Features = Bottom_Picking[['open', 'high', 'low', 'close','volume','Relative Strength', 'RSI 14 ', 'MA 20', 'MA 50', 'MA 100','Trade Status']]
Trend_Following_Features = Trend_Following[['open', 'high', 'low', 'close','volume','Relative Strength', 'RSI 14 ', 'MA 20', 'MA 50', 'MA 100','Trade Status']]
```

```python
print(Bottom_Picking_Features.shape)
print(Trend_Following_Features.shape)
```

# Apply Clustering Algorithm on each Dataset

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

## Clustering for Trend Following

```python
Trend_Following_Features.fillna(0,inplace=True)
```

```python
kmeans.fit(Trend_Following_Features)
```

```python
Trend_Following_Labels = kmeans.labels_
```

```python
Trend_Following['Cluster'] = Trend_Following_Labels
```

## Clustering for Bottom Picking

```python
Bottom_Picking_Features.fillna(0,inplace=True)
```

```python
kmeans.fit(Bottom_Picking_Features)
```

```python
Bottom_Picking_Labels = kmeans.labels_
```

```python
Bottom_Picking['Cluster'] = Bottom_Picking_Labels
```
