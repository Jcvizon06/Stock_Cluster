```python
'''
Importing all the necessary libraries
'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Stock Features.csv')
```

```python
'''
Replace Null data with 0
'''
df.fillna(0, inplace=True)
```

```python
df.head(20)
```

```python
def ma_snap(bol):
    if bol == True:
        return 1
    else:
        return 0
```

```python
df['MA Snap'] = df['MA 20 days'] > df['close']
```

```python
df['MA Snap'] = df['MA Snap'].apply(ma_snap)
```

```python
df['MA Snap'].unique()
```

```python
df.columns
```

```python
df.dtypes
```

```python
'''
Features are the numeric data only stock code and date not included
'''
features = df[['open', 'high', 'low', 'close', 'volume',
       'MA 20 days', 'MA 50 days', 'MA 100 days', 'RSI 14 days', 'MA Snap']]
```

# K-Means Clusting Algorithm


For video refference how it works please click on the link below


https://www.youtube.com/watch?v=3vHqmPF4VBA

```python
from sklearn.cluster import KMeans
```

```python
'''
Instantiate K_means and define 3 cluster
'''
k_means = KMeans(n_clusters=3)
```

```python
'''
fit it on the dataset
'''
k_means.fit_transform(features)
```

```python
k_means.labels_
```

```python
df['Cluster'] = k_means.labels_
```

```python
cluster_0 = df[df['Cluster'] == 0]
```

```python
cluster_1 = df[df['Cluster'] == 1]
```

```python
cluster_2 = df[df['Cluster'] == 2]
```
