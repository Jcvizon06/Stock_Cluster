```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.fillna(0, inplace=True)
```

```python
df.columns
```

```python
features = df[['open', 'high', 'low', 'close', 'volume',
       'Upward Movement', 'Downward Movement', 'Ave Upward', 'Ave Movement',
       'Relative Strength', 'RSI 14 ', 'MA 20', 'MA 50', 'MA 100']]
```

```python
features.head(10)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(features)
```

```python
kmeans.labels_
```

```python
df['Cluster'] = kmeans.labels_
```

```python
df['Cluster'].unique()
```

```python
cleaned_data_cluster_0 = df[df['Cluster'] == 0]
```

```python
cleaned_data_cluster_1 = df[df['Cluster'] == 1]
```

```python
cleaned_data_cluster_2 = df[df['Cluster'] == 2]
```
